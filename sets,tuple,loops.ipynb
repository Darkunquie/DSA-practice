```python
DICT ={
    123: [1,2,3],
    'greeting':'hi',
     'age':11

}
print(123 in DICT)
print('greeting' in DICT)
print('greeting' not in DICT)
```

```python
#KEYS
DICT ={
    123: [1,2,3],
    'greeting':'hi',
     'age':11

}
print(DICT.keys())
```

```python
#ITEMS
DICT ={
    123: [1,2,3],
    'greeting':'hi',
     'age':11

}
print(DICT.items())
```

```python
#POP
DICT ={
    123: [1,2,3],
    'greeting':'hi',
     'age':11

}
print(DICT.pop(123))
print(DICT)
```

```python
#update
DICT ={
    123: [1,2,3],
    'greeting':'hi',
     'age':11

}
print(DICT.update({'age':55}))
print(DICT)
```
