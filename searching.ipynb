##Linear Search Algorithm

##When to use Linear Search Algorithm?
When we are dealing with a small dataset.

When you are searching for a dataset stored in contiguous memory.

```python
# Python3 code to linearly search x in arr[].


def search(arr, N, x):

    for i in range(0, N):
        if (arr[i] == x):
          print(arr[i])
          return i

    return -1

if __name__ == "__main__":
    arr = [2, 3, 4, 10, 40]
    x = 40
    N = len(arr)

    # Function call
    result = search(arr, N, x)
    if(result == -1):
        print("Element is not present in array")
    else:
        print("Element is present at index", result)
```

```python
# Python3 code to implement iterative Binary
# Search.


# It returns location of x in given array arr
def binarySearch(arr, low, high, x):

    while low <= high:

        mid = low + (high - low) // 2

        # Check if x is present at mid
        if arr[mid] == x:
            return mid

        # If x is greater, ignore left half
        elif arr[mid] < x:
            low = mid + 1

        # If x is smaller, ignore right half
        else:
            high = mid - 1

    # If we reach here, then the element
    # was not present
    return -1


# Driver Code
if __name__ == '__main__':
    arr = [2, 3, 4, 10, 40]
    x = 10

    # Function call
    result = binarySearch(arr, 0, len(arr)-1, x)
    if result != -1:
        print("Element is present at index", result)
    else:
        print("Element is not present in array")
```

```python
# Python program to find the second largest element in the array
# using one traversal

# function to find the second largest element in the array
def getSecondLargest(arr):
    n = len(arr)

    largest = -1
    secondLargest = -1

    # finding the second largest element
    for i in range(n):

        # If arr[i] > largest, update second largest with
        # largest and largest with arr[i]
        if arr[i] > largest:
            secondLargest = largest
            largest = arr[i]

        # If arr[i] < largest and arr[i] > second largest,
        # update second largest with arr[i]
        elif arr[i] < largest and arr[i] > secondLargest:
            secondLargest = arr[i]

    return secondLargest

if __name__ == "__main__":
    arr = [12, 35, 1, 10, 34, 1]
    print(getSecondLargest(arr))
```

```python
def find_first_repeating_element(arr):


  seen = set()  # Use a set for efficient lookups
  for element in arr:
    if element in seen:
      return element  # Return immediately when a repeating element is found
    seen.add(element)  # Add the element to the set if it's not already seen
  return None  # Return None if no repeating element is found

if __name__ == "__main__":
  arr = [10, 5, 3, 4, 3, 5, 6]
  repeating_element = find_first_repeating_element(arr)
  if repeating_element:
    print(f"The first repeating element is: {repeating_element}")
  else:
    print("No repeating element found.")
```

```python
def min_abs_sum_pair(arr):
  res=arr[0]+arr[1]
  for i in range(len(arr)-1):
    for j in range(i+1,len(arr)):
      res=min(res,abs(arr[i]+arr[j]))
  return res


arr = [1, 60, -10, 70, -80, 85]
print(min_abs_sum_pair(arr))
```

```python
def find_pair_with_difference(arr, diff):
    """
    Finds a pair of numbers in the given list that have the specified difference.

    Args:
        arr: The list of numbers.
        diff: The target difference.

    Returns:
        A tuple containing the pair of numbers, or None if no such pair is found.
    """
    num_set = set(arr)  # Convert the list to a set for efficient lookups

    for num in arr:
        complement = num - diff  # Calculate the complement needed for the target difference
        if complement in num_set:  # Check if the complement exists in the set
            return (num, complement)  # If found, return the pair

    return None  # If no pair is found, return None

if __name__ == "__main__":
    arr = [5, 20, 3, 2, 50, 80]
    diff = 78
    result = find_pair_with_difference(arr, diff)
    if result:
        print("Pair Found:", result)
    else:
        print("No Such Pair")
```

Find k largest elements in an array

```python
# Python program to find k largest elements in an
# array using sorting

def kLargest(arr, k):

    # sort the given array in descending order
    arr.sort(reverse=True)

    # store the first k elements in result list
    res = arr[:k]
    return res

if __name__ == "__main__":
    arr = [1, 23, 12, 9, 30, 2, 50]
    k = 3
    res = kLargest(arr, k)
    print(' '.join(map(str, res)))

```

```python
Input:  mat ={{10, 20, 30, 40},
                        {15, 25, 35, 45},
                        {24, 29, 37, 48},
                        {32, 33, 39, 50 }}
K = 3
Output: 20
Explanation: The 3rd smallest element is 20



Input:  mat ={{10, 20, 30, 40},
                        {15, 25, 35, 45},
                        {24, 29, 37, 48},
                        {32, 33, 39, 50 }}
K = 7
Output: 30
Explanation: The 7th smallest element is 30
```

```python
def tri(arr):
  res=[]
  n=len(arr)
  for i in range(n-2):
    for j in range(i+1,n-1):
      for k in range(j+1,n):
        if arr[i]+arr[j]+arr[k] == 0:
          res.append((arr[i],arr[j],arr[k]))
  return res
if __name__=="__main__":
  arr=[-1,0,1,2,-1,-4]
  res=tri(arr)
  for i in res:
    print(i[0],i[1],i[2])
```
