```python
score = float(input("Enter the score: "))

if score >= 90:
    print("A")
elif score >= 80:
    print("B")
elif score >= 70:
    print("C")
elif score >= 60:
    print("D")
else:
    print("F")
```

```python
grades = {}
print("Initial grades:", grades)
```

```python
n = "Ana"
s = 90
grades[n] = s

n = "Ben"
s = 85
grades[n] = s

print("Grades after adding:", grades)
```

```python
up_n = "Ana"
new_s = 92

if up_n in grades:
    grades[up_n] = new_s
    print(f"Updated {up_n}'s score to {new_s}.")
else:
    print(f"Student {up_n} not found.")

print("Grades after update:", grades)
```

```python
print("\n--- All Grades ---")
if grades:
    for n, s in grades.items():
        print(f"{n}: {s}")
else:
    print("No grades yet.")
print("------------------")
```

```python
f_name = "simple_file.txt"

text_to_write = "This is a very simple line of text."

with open(f_name, 'w') as f:
    f.write(text_to_write)
print(f"Wrote to '{f_name}'.")

with open(f_name, 'r') as f:
    read_text = f.read()
print(f"Read from '{f_name}':\n{read_text}")
```
